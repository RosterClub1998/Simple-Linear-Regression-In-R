```r
#Loading ggplot2 package for creating attractive visualisations in R
library(ggplot2)
```

```r
#Importing data set from url
dataset <- read.csv(url("http://bit.ly/w-data"))
```

```r
#To get a view of data that is imported use print() command or simply type object_name and hit enter
print(dataset)
```

```r
#To get first 5 records use head Command
head(dataset,5)
```

```r
#Used ggplot() command to create viz..
plot<-ggplot(dataset,aes(x=Hours,y=Scores)) + geom_point()+stat_smooth(method = "lm", col = "dodgerblue3") +ggtitle("Linear Model fitted to data")
```

```r
plot
```

```r
#Used lm() function to cread linear model
r<-lm(Scores~Hours,data=dataset)
```

```r
r
```

```r
#Used cor() function to get corelation between dependent and independent variable
cor(dataset$Hours,dataset$Scores)
```

```r
summary(r)
```

```r
#Used names() to access regression objects
names(r)
```

```r
#fitted is a generic function which extracts fitted values from objects returned by modeling functions.
fitted(r)
```

```r
#fit: the predicted sale values for the three new advertising budget
#lwr and upr: the lower and the upper confidence limits for the expected values, respectively. By default the function produces the 95% confidence limits.
predict(r,data.frame(Hours=9.25),interval="confidence",level=0.9)
```
